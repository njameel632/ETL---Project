```python
# Salaries By Region and College Type.
# College Types are categorized into Engineering, State, Party, Liberal Arts, Ive League
```

```python
## Reading 1st Dataset - Salaries By College

salaries_college = "Resources/Salaries-College.csv"
salaries_college_data = pd.read_csv(salaries_college)
salaries_college_data.head()
```

```python
## Dropping Columns to Inlcude only Starting & Mid-Career Salaries

new_salaries_college = salaries_college_data[['School Name','School Type','Starting Median Salary']]
new_salaries_college = new_salaries_college.rename(columns={'School Name':'School_name','School Type': 'School Known For', 'Starting Median Salary': 'Starting_Median_Salary'})
new_salaries_college.head()
```

```python
## Schools with Highest Median salary

highest_salary = new_salaries_college.sort_values(by='Starting_Median_Salary', ascending=False)
highest_salary.head(10)
```

```python
# Grouping College By Types

college_type = new_salaries_college.groupby(['School Known For']).count()['School_name']
college_type
```

```python
## Loading 2nd Dataset Salaries By region

salaries_region = "Resources/Salaries-Region.csv"
salaries_region_data = pd.read_csv(salaries_region)
salaries_region_data.head()
```

```python
## Replacing NaN with N/A // Dropping Columns to Inlcude only Starung & Mid-Career Salary

salaries_region_data.fillna('N/A')
new_salaries_region = salaries_region_data[['School Name','Region','Starting Median Salary']]
new_salaries_region = new_salaries_region.rename(columns={'School Name':'School_name', 'Starting Median Salary':'Starting_Median_Salary'})
new_salaries_region.head()
```

```python
## Merging 2 datasets

salaries_college_region = pd.merge(new_salaries_college, new_salaries_region, on='School_name', how='left')
salaries_college_region = salaries_college_region[['School_name','Starting_Median_Salary_x','Region']]
salaries_college_region = salaries_college_region.rename(columns={'Starting_Median_Salary_x':'Starting_Median_Salary'})
salaries_college_region.head()
```

```python
## Finding highest salary by region

highest_salary_region = salaries_college_region.sort_values(by='Starting_Median_Salary', ascending=False)
highest_salary_region.head(10)
```

```python
# Lowest Salary By region
highest_salary_region = salaries_college_region.sort_values(by='Starting_Median_Salary', ascending=True)
highest_salary_region.head(10)
```

```python
## Loading Tuition Dataset to get tuition cost for universities

tuition_cost ="Resources/tuition_cost.csv"
tuition_cost_data = pd.read_csv(tuition_cost)
tuition_cost_data.head()
tuition_cost_data.columns
```

```python
## Dropping unwanted columns and renaming columns to merge

tuition_cost_data = tuition_cost_data[['name','type','degree_length','in_state_total','out_of_state_total']]
tuition_cost_data
```

```python
## Dropping all schools who have a degree length of 2 years

tuition_cost_data = tuition_cost_data.loc[tuition_cost_data['degree_length']=='4 Year']
```

```python
tuition_cost_data = tuition_cost_data.rename(columns={'name':'School_name','type':'School_Type','degree_length':'Degree_Length','in_state_total':'Tuition_Cost_Per_Year'})
tuition_cost_data = tuition_cost_data[['School_name','School_Type','Degree_Length','Tuition_Cost_Per_Year']]
tuition_cost_data["Tuition_Cost_Per_Year"] = tuition_cost_data["Tuition_Cost_Per_Year"].astype(float).map("${:,.2f}".format)
```
