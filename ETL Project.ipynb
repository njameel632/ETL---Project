```python
# Salaries By Region and College Type.
# College Types are categorized into Engineering, State, Party, Liberal Arts, Ive League
```

```python
## Reading 1st Dataset - Salaries By College

salaries_college = "Resources/Salaries-College.csv"
salaries_college_data = pd.read_csv(salaries_college)
salaries_college_data.head()
```

```python
## Dropping Columns to Inlcude only Starting & Mid-Career Salaries

new_salaries_college = salaries_college_data[['School Name','School Type','Starting Median Salary']]
new_salaries_college = new_salaries_college.rename(columns={'School Type': 'School Known For'})
new_salaries_college.head()
```

```python
# Grouping College By Types

college_type = new_salaries_college.groupby(['School Known For']).count()['School Name']
college_type
```

```python
## Loading 2nd Dataset Salaries By region

salaries_region = "Resources/Salaries-Region.csv"
salaries_region_data = pd.read_csv(salaries_region)
salaries_region_data.head()
```

```python
## Replacing NaN with N/A // Dropping Columns to Inlcude only Starung & Mid-Career Salary

salaries_region_data.fillna('N/A')
new_salaries_region = salaries_region_data[['School Name','Region','Starting Median Salary']]
new_salaries_region
```

```python
## Merging 2 Datasets using key as School Name

salaries_college_region = pd.merge(new_salaries_college, new_salaries_region, on='School Name', how='left')
salaries_college_region = salaries_college_region[['School Name', 'Starting Median Salary_x', 'Region']]
salaries_college_region = salaries_college_region.rename(columns={'Starting Median Salary_x':'Starting Median Salary'})
salaries_college_region.head()
```

```python
## Loading Tuition Dataset to get tuition cost for universities

tuition_cost ="Resources/tuition_cost.csv"
tuition_cost_data = pd.read_csv(tuition_cost)
tuition_cost_data.head()
tuition_cost_data.columns
```

```python
## Dropping unwanted columns and renaming columns to merge

tuition_cost_data = tuition_cost_data[['name','type','degree_length','in_state_total','out_of_state_total']]
tuition_cost_data
```

```python
## Dropping all schools who have a degree length of 2 years

tuition_cost_data = tuition_cost_data.loc[tuition_cost_data['degree_length']=='4 Year']
tuition_cost_data.count()
```

```python
tuition_cost_data = tuition_cost_data.rename(columns={'name':'School Name','type':'School Type','degree_length':'Degree Length','in_state_total':'Tuition Cost Per Year'})
tuition_cost_data = tuition_cost_data[['School Name','School Type','Degree Length','Tuition Cost Per Year']]
tuition_cost_data.head()
```

```python
rds_connection_string = "postgres:3C:2E:FF:1F:AA:0F@localhost:5433/ETLProject"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
## Pushing data into Postgresql

new_salaries_region.to_sql(name='salaries_data', con=engine, if_exists='append', index=False)
tuition_cost_data.to_sql(name='tuition_data', con=engine, if_exists='append', index=False)
```
